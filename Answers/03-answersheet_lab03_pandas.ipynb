```python
df.groupby(['work_year'])[['salary_in_usd', 'job_title']].apply(lambda x: x.sort_values(
                                               by='salary_in_usd', ascending=False)[:3])
```

```python
df.groupby(['work_year'])[['salary', 'job_title']].apply(lambda x: x.sort_values(
                                                                   by='salary', ascending=False).head(3))
```

```python
work_year = df['work_year'].unique()
work_year
```

```python
df.groupby('work_year')[['job_title', 'salary_in_usd']].apply(lambda x: (x.groupby(
    'job_title')[['salary_in_usd']].mean()).sort_values(by='salary_in_usd', ascending=False)[:3])
```

```python
df.groupby(['work_year', 'job_title'])[['salary_in_usd']].groups
```

```python
df.groupby(['work_year', 'job_title'], as_index=False)[['salary_in_usd']].mean()
```

```python
df.groupby(['work_year', 'job_title'], as_index=False)[['salary_in_usd']].mean().groupby(
    'work_year').apply(lambda x: x.sort_values(by='salary_in_usd', ascending=False)[:3])
```
